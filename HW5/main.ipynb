```python
import numpy as np
from nn import NN
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# Unit = 3
model = NN(3)
```

```python
data = pd.read_csv('xor.csv', header=None).to_numpy()
```

```python
xs = data[:,:2]
ys = data[:,2:]
```

```python
model.build(2)
```

```python
loss = []
for _ in range(3):  #1000 steps
    # 50 steps in an epoch
    for i in range(len(data)):
        loss.append(model.train(xs[i], ys[i], lr=0.1))
plt.plot(np.arange(len(loss)), loss)
plt.xlabel('Iterations')
plt.ylabel('Cross Entropy')
plt.title('Loss vs. Iterations')
plt.show()
```
